# Phase 3: Subject-to-Template Evaluation

This notebook compares voxel overlap measures against AFID-based metrics for evaluating spatial correspondence. The OASIS-1 dataset from from PHASE2 was processed using the Ants-based T1-to-MNI (`MNI152NLin2009bAsym`) registration workflow built-in to fMRIPrep.

```r
# initialize libraries
library(plyr)
library(dplyr)
library(digest)
library(reshape2)
library(ggplot2)
library(ggpubr)
```

```r
# useful functions

# calculate the distance between two sets of coordinates
dist3D <- function(coord1, coord2) { # vector X,Y,Z
        xdist <- coord1[1] - coord2[1] # could also write as coord1$X, etc.
        ydist <- coord1[2] - coord2[2]
        zdist <- coord1[3] - coord2[3]
        euclidean_dist <- sqrt(xdist^2+ydist^2+zdist^2)
        return(as.numeric(unlist(euclidean_dist)))
}

# calculate the pairwise distance between an array of 3D coordinates
pairwise_dist3D <- function(temp_coords) { # labeled X,Y,Z
        N <- length(temp_coords$X)
        dist_vec <- rep(0,N) # create vector
        sum_dist <- 0 # initialize to zero
        count <- 0
        for (i in 1:(N-1)) {
                for (j in (i+1):N) {
                        if (i != j) {
                                count <- count + 1
                                first_coord <- temp_coords[i,]
                                second_coord <- temp_coords[j,]
                                curr_dist <- dist3D(first_coord, second_coord)
                                sum_dist <- sum_dist + curr_dist
                                dist_vec[count] <- curr_dist
                        }
                }
        }
        return(c(as.numeric(unlist(mean(dist_vec))),as.numeric(unlist(sd(dist_vec)))))
}
```

## ROI Overlap

Values for pallidum, striatum, and thalamus.

```r
# initialize variables and calculate jaccard/kappa
df_afid <- read.table('~/GitHub/afids-analysis/etc/afids.csv', sep=",", header=TRUE)
df_rois_lin <- read.table('~/GitHub/afids-analysis/data/PHASE3_output_maps/PHASE3_output_rois_lin.csv', sep=",", header=TRUE)
df_rois_lin$xfm <- 'lin'
df_rois_nlin <- read.table('~/GitHub/afids-analysis/data/PHASE3_output_maps/PHASE3_output_rois_nlin.csv', sep=",", header=TRUE)
df_rois_nlin$xfm <- 'nlin'
df_rois <- rbind(df_rois_lin,df_rois_nlin)
df_rois$subject <- gsub("sub-", "OAS1_", df_rois$subject)

# calculate jaccard
df_rois$jaccard = df_rois$intersection / df_rois$union

# calculate kappa (i.e. dice)
df_rois$kappa = (2*df_rois$intersection) / (df_rois$id_vol + df_rois$truth_vol)
```

```r
# dplyr summary table creation

#example1 <- wilcox.test( 
#    subset(df_rois, xfm == 'lin' & roi == 'pallidum' & side == 'left')$jaccard,
#    subset(df_rois, xfm == 'nlin' & roi == 'pallidum' & side == 'left')$jaccard
#)
#names(example1)
#example1

roi_res_jaccard <- df_rois %>% group_by(roi,side) %>%
       do(w = wilcox.test(jaccard~xfm, data=.)) %>% 
       dplyr::summarise(roi, side, W_pval_jaccard = w$p.value)
roi_res_kappa <- df_rois %>% group_by(roi,side) %>%
       do(w = wilcox.test(jaccard~xfm, data=.)) %>% 
       dplyr::summarise(roi, side, W_pval_kappa = w$p.value)

df_overlap <- df_rois[,c(1:3,8:10)] # simplify for creation of the summary table
summary_roi_overlap <- subset(df_overlap, xfm == 'lin')
summary_roi_overlap <- merge( summary_roi_overlap, subset(df_overlap, xfm == 'nlin'), by = c('subject','roi','side') )
names(summary_roi_overlap)[5] <- 'jaccard_lin'
names(summary_roi_overlap)[6] <- 'kappa_lin'
names(summary_roi_overlap)[8] <- 'jaccard_nlin'
names(summary_roi_overlap)[9] <- 'kappa_nlin'

summary_roi_overlap <- ddply(summary_roi_overlap, .(roi,side), summarize,
            mean_jaccard_lin = mean(jaccard_lin), sd_jaccard_lin = sd(jaccard_lin),
            mean_kappa_lin= mean(kappa_lin), sd_kappa_lin = sd(kappa_lin),
            mean_jaccard_nlin = mean(jaccard_nlin), sd_jaccard_nlin = sd(jaccard_nlin),
            mean_kappa_nlin= mean(kappa_nlin), sd_kappa_nlin = sd(kappa_nlin)        
)

summary_roi_overlap <- merge( summary_roi_overlap, roi_res_jaccard, by = c('roi','side'))
summary_roi_overlap <- merge( summary_roi_overlap, roi_res_kappa, by = c('roi','side'))

# FDR correction
summary_roi_overlap$W_pval_jaccard_adjusted <- p.adjust(summary_roi_overlap$W_pval_jaccard, "fdr")
summary_roi_overlap$W_pval_jaccard_significant <- (summary_roi_overlap$W_pval_jaccard_adjusted < 0.05)
summary_roi_overlap$W_pval_kappa_adjusted <- p.adjust(summary_roi_overlap$W_pval_kappa, "fdr")
summary_roi_overlap$W_pval_kappa_significant <- (summary_roi_overlap$W_pval_kappa_adjusted < 0.05)

# all results statistically significant even after fdr correction

# round and rearrange for display
output_table_roi_overlap <- summary_roi_overlap[,c(1,2)]

output_table_roi_overlap$jaccard_lin <- paste0( sprintf( "%.2f", round(summary_roi_overlap$mean_jaccard_lin,2)), '±', sprintf( "%.2f", round(summary_roi_overlap$sd_jaccard_lin,2)) )
output_table_roi_overlap$jaccard_nlin <- paste0( sprintf( "%.2f", round(summary_roi_overlap$mean_jaccard_nlin,2)), '±', sprintf( "%.2f", round(summary_roi_overlap$sd_jaccard_nlin,2)) )
#output_table_roi_overlap$jaccard_qval <- format.pval(summary_roi_overlap$W_pval_jaccard_adjusted, digits = 2, eps = 0.001, nsmall = 3)
output_table_roi_overlap$jaccard_lin_vs_nlin <- ifelse(summary_roi_overlap$W_pval_jaccard_adjusted, '*', '')

output_table_roi_overlap$kappa_lin <- paste0( sprintf( "%.2f", round(summary_roi_overlap$mean_kappa_lin,2)), '±', sprintf( "%.2f", round(summary_roi_overlap$sd_kappa_lin,2)) )
output_table_roi_overlap$kappa_nlin <- paste0( sprintf( "%.2f", round(summary_roi_overlap$mean_kappa_nlin,2)), '±', sprintf( "%.2f", round(summary_roi_overlap$sd_kappa_nlin,2)) )
#output_table_roi_overlap$kappa_qval <- format.pval(summary_roi_overlap$W_pval_kappa_adjusted, digits = 2, eps = 0.001, nsmall = 3)
output_table_roi_overlap$kappa_lin_vs_nlin <- ifelse(summary_roi_overlap$W_pval_kappa_adjusted, '*', '')

output_table_roi_overlap

#write.table(output_table_roi_overlap, file = "~/GitHub/afids-analysis/data/output_tables/PHASE3_roi_overlap.csv", row.names = FALSE, quote = FALSE, sep = ",")
```

# AFRE

Anatomical FRE is evaluated here as a metric for looking at the spatial correspondence between images. Here they are summarized globally, for each AFID, and for each subject. Qualitatively, the ventricles are misaligned for OAS1_0109 which accounts for the maximally error observed in this analysis of > 30 mm AFRE.

```r
# load the template: ground truth placements for ICBM2009b-Asym == ICBM2009c-Asym
df_ICBM2009bAsym_mean <- data.frame(fid=integer(),X=double(),Y=double(),Z=double(),stringsAsFactors = FALSE)
ICBM2009bAsym_file <- "~/GitHub/afids-analysis/data/PHASE1_output_afid_postQC/MNI152NLin2009bAsym_MEAN_postQC.fcsv"

curr_template <- read.table(ICBM2009bAsym_file, header=FALSE, sep=",")
df_template <- data.frame(fid = 1:length(curr_template$V1))
df_template <- cbind(df_template,X=curr_template[2],Y=curr_template[3],Z=curr_template[4])
df_template <- plyr::rename(df_template, c("V2"="X","V3"="Y","V4"="Z"))
df_ICBM2009bAsym_mean <- df_template
#df_ICBM2009bAsym_mean
```

## Nonlinear Transform Results

```r
# also load all of the individual transformed placements for all 30 subjects
# compute AFRE
df_subjects_nlin <- data.frame(uid=integer(), fid=integer(),X=double(),Y=double(),Z=double(),subject=factor(),
                        subject=factor(),mri_session=factor(),mri_type=factor(),session=integer(),date=integer(),
                        name=character(),description=character(),stringsAsFactors = FALSE)

setwd('~/GitHub/afids-analysis/data/PHASE3_output_afid_nlin/')
csv_files <- list.files(".", "*.fcsv")

for (i in 1:length(csv_files)) {
        curr_split <- unlist(strsplit(csv_files[i],"_"))
        if (length(curr_split)>1) { # extract name and session data
                subject_id <- paste0( 'OAS1_', curr_split[2])
                subject_mri_type <- curr_split[3]
        }
        
        curr_subject <- read.table(csv_files[i], header=FALSE, sep=",")
        df_subject <- data.frame(fid = 1:length(curr_subject$V1))
        
        df_subject <- cbind(df_subject,subject=subject_id,
                            name=curr_subject[12], description=curr_subject[13],
                            X=curr_subject[2],Y=curr_subject[3],Z=curr_subject[4])
        
        df_subject <- plyr::rename(df_subject, c("V2"="X","V3"="Y","V4"="Z","V12"="name","V13"="description"))
        df_subjects_nlin <- rbind(df_subjects_nlin,df_subject)
}
```

```r
# Compute the AFRE now
df_subjects_nlin <- merge(df_subjects_nlin, df_ICBM2009bAsym_mean, by = 'fid')
df_subjects_nlin$AFRE <- dist3D(df_subjects_nlin[,5:7],df_subjects_nlin[,8:10])

sprintf( "Total: %.2f +/- %.2f mm; Range: %.2f-%.2f",
         mean(df_subjects_nlin$AFRE), sd(df_subjects_nlin$AFRE), min(df_subjects_nlin$AFRE), max(df_subjects_nlin$AFRE) )

summary_afids_AFRE_nlin_df <- ddply(df_subjects_nlin, .(fid), summarize, mean_AFRE = mean(AFRE), sd_AFRE = sd(AFRE), max_AFRE = max(AFRE))
summary_afids_AFRE_nlin_df[,-1] <- round(summary_afids_AFRE_nlin_df[,-1],2)
summary_afids_AFRE_nlin_df <- merge(df_afid,summary_afids_AFRE_nlin_df,by.x='name',by.y='fid')
#summary_afids_AFRE_df

summary_subjects_AFRE_nlin_df <- ddply(df_subjects_nlin, .(subject), summarize, mean_AFRE = mean(AFRE), sd_AFRE = sd(AFRE), max_AFRE = max(AFRE))
summary_subjects_AFRE_nlin_df[,-1] <- round(summary_subjects_AFRE_nlin_df[,-1],2)
#summary_subjects_AFRE_nlin_df

mean(summary_subjects_AFRE_nlin_df$max_AFRE)
```

## Correlation between AFRE and voxel overlap

A weak negative correlation was found between AFRE and standard voxel overlap measures at the global dataset level and for each specific ROI in isolation.

```r
# note that Jaccard and Kappa are fully correlated and thus the kendall tau values are the same.
cor.test(compare_overlap_AFRE_nlin$jaccard,compare_overlap_AFRE_nlin$AFRE,method='kendall') # jaccard and AFRE
cor.test(compare_overlap_AFRE_nlin$kappa,compare_overlap_AFRE_nlin$AFRE,method='kendall') # kappa and AFRE
cor.test(compare_overlap_AFRE_nlin$jaccard,compare_overlap_AFRE_nlin$kappa,method='kendall') # jaccard and kappa

summary_AFRE_vs_ROI <- ddply(compare_overlap_AFRE_nlin,.(roi,side),summarize,AFRE_jaccard=cor.test(jaccard,AFRE,method = 'kendall')$estimate,AFRE_jaccard_pval=cor.test(jaccard,AFRE,method = 'kendall')$p.value,AFRE_kappa=cor.test(kappa,AFRE,method = 'kendall')$estimate,AFRE_kappa_pval=cor.test(kappa,AFRE,method = 'kendall')$p.value)

# FDR correction
summary_AFRE_vs_ROI$AFRE_jaccard_pval_adjusted <- p.adjust(summary_AFRE_vs_ROI$AFRE_jaccard_pval, "fdr")
summary_AFRE_vs_ROI$AFRE_jaccard_pval_significant <- (summary_AFRE_vs_ROI$AFRE_jaccard_pval_adjusted < 0.05)

summary_AFRE_vs_ROI$AFRE_kappa_pval_adjusted <- p.adjust(summary_AFRE_vs_ROI$AFRE_kappa_pval, "fdr")
summary_AFRE_vs_ROI$AFRE_kappa_pval_significant <- (summary_AFRE_vs_ROI$AFRE_kappa_pval_adjusted < 0.05)

summary_AFRE_vs_ROI
```

## For each AFID and ROI

No correlation between voxel overlap measures and individual AFID AFREs were identified. However, plotting of voxel overlap against individual AFREs demonstrate the added sensitivity to misregistration when looking at individual AFID plots along the y-axis.

# AFLE versus AFRE

In this section, we examine whether AFLE and AFRE are correlated and establish a baseline for when AFRE should be considered beyond placement-related (AFLE) error. The first vertical line is the mean AFLE for OASIS-1 subjects. Second is 1 s.d., third is 2 s.d.

A positive correlation between AFLE and AFRE was found to be statistically significant although the actual effect size of the correlation was small.

```r
# AFLE versus AFRE
df_AFLE <- read.table('~/GitHub/afids-analysis/data/PHASE2_output_afid_postQC/PHASE2_subject_validation_AFLE.csv', sep=",", header=TRUE)
AFLE_vs_AFRE <- merge(df_AFLE, df_subjects, by = c('subject','fid'))

# simplest to just use 1 mm
afle_mean <- 1 # simplest to just use 1 mm
afle_sd <- 1 # simplest to just use 1 mm

AFLE_vs_AFRE <- subset( AFLE_vs_AFRE, xfm == 'nlin' ) # only interested in nlin group for this analysis

# local AFRE
# thresholding based on AFLE mean and sd across all fids; may be worth thresholding based on mean and sd of each specific fid
AFRE_outliers <- subset(df_subjects, xfm == 'nlin' & (AFRE > afle_mean + afle_sd*2))
#AFRE_outliers

sprintf( "Num Outlier AFIDs (> 2 s.d. above mean AFLE): %d/%d (%.2f%%)",
        length(AFRE_outliers$fid), length(AFLE_vs_AFRE$fid), 100*length(AFRE_outliers$fid)/length(AFLE_vs_AFRE$fid) )

sprintf( "Num Unique Outlier AFIDs (> 2 s.d. above mean AFLE): %d/%d (%.2f%%)",
        length(unique(AFRE_outliers$fid)), 32, 100*length(unique(AFRE_outliers$fid))/32 )

cor.test(AFLE_vs_AFRE$AFRE, AFLE_vs_AFRE$mean_AFLE, method='kendall')
```

```r
## plot of AFLE and AFRE
ggplot(AFLE_vs_AFRE, aes(AFRE, mean_AFLE)) +
    geom_point(alpha = 0.6) +
    geom_vline(xintercept=afle_mean) +
    geom_vline(xintercept=afle_mean+afle_sd) +
    geom_vline(xintercept=afle_mean+2*afle_sd) +
    scale_x_continuous(breaks=c(seq(0,10,1))) +
    scale_y_continuous(breaks=c(seq(0,10,1))) +
    coord_cartesian(xlim=c(0, 10), ylim=c(0, 10)) +
    labs(x = "AFRE (mm)", y = "AFLE (mm)") +
    theme_bw()
```

# Secondary Analyses

We evaluated whether there was any evidence of an effect of demographics (e.g. age of the participants being registered) on AFRE. Age resulted in a global AFRE change of 0.0075 mm/year (i.e. a small but statistically significant effect). No specific AFIDs were found to contribute to this age-related AFRE change after multiple comparisons correction.

```r
df_OAS1 <- read.table('~/GitHub/landmarkprotocol/etc/OAS1_demographics.csv', sep=",", header=TRUE)
# include only MR1s
df_OAS1 <- subset(df_OAS1, session == 'MR1')

# also look at age and gender of subject and impact on AFRE
df_demographics <- merge(df_subjects, df_OAS1, by = 'subject')

l <- lm(AFRE ~ age, data = df_demographics)
s <- summary(l) # for age, the estimate was positive: 0.0075
round(cbind(l$coeff,s$coefficients[,4]),4) # first column is the effect, second column is the pval
```

```r
# Did AFRE change with the age of the participant for specific AFIDs?
# create dataframe for linear model for each rater and p-values
models = dlply(df_demographics, .(fid), lm, formula = AFRE ~ age)
# also extract p-values for intercept and session
qual <- laply(models, function(mod) summary(mod)$coefficients[,4])

coefs = ldply(models, coef)
summary_aging_afids <- cbind(coefs,qual)
summary_aging_afids <- summary_aging_afids[,c(1,2,4,3,5)]
names(summary_aging_afids)[c(3,5)] <- c('pval_(Intercept)','pval_age')

# FDR correction
summary_aging_afids$pval_age_adjusted <- p.adjust(summary_aging_afids$pval_age, "fdr")
summary_aging_afids$pval_age_significant <- (summary_aging_afids$pval_age_adjusted < 0.05)

# Round and display
summary_aging_afids[,c(2,4)] <- round( summary_aging_afids[,c(2,4)], 2)
summary_aging_afids[,c(3,5,6)] <- round( summary_aging_afids[,c(3,5,6)], 2)              
summary_aging_afids
```
